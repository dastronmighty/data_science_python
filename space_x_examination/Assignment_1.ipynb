# Examination of SpaceX data from r/spacex

```python
import requests
import csv
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Data identification:

###### I will be using an opensource SpaceX data API I chose this as I am very interested in SpaceX the Company and wanted to look at something I was interested in. This is an opensource API maintsained by the r/spacex community on Reddit I will be looking at all of SpaceX's looking at past launches. The API has lots of data so I am limiting the features I want. I wil be looking at all hostorical falcon9 launches and falcon 9 payloads.

###### I used postman (https://www.getpostman.com) to check the data first to identify the features I wanted

```python
API_URL = "https://api.spacexdata.com/v3/launches"
SPACEX_DATA_FILE = r'./SpaceX_Historical_Launches.csv'
```

## 2. Data Collection 

```python
headers = [] 
# our headers are going to be our features
headers.append('flight_number')
headers.append('launch_year')
headers.append('launch_date')
headers.append('launch_success')
headers.append('times_flown')
headers.append('payload_mass')
headers.append('periapsis')
headers.append('apoapsis')
headers.append('inclination')
headers.append('orbital_period')
headers.append('reused')
headers.append('landing_intent')
headers.append('land_success')
headers.append('site_name')
```

the headers here are what will be the headings in our comma seperated value (CSV) file 

```python
def create_csv_file(file, header):
    with open(file, "w",newline='') as write_file:
        csv_file = csv.writer(write_file) #CREATE A NEW CSV FILE
        csv_file.writerow(header) #write our features as the headers
    write_file.close()

def get_data_from_url(url): 
    response = requests.get(url) #send a get request to the specified URL 
    res_body = response.json() #extract the json version of the body of the response 
    return res_body

def write_json_to_csv(file, data):
    
     with open(file, "a", newline='') as write_file: # open our csv in append mode
        csv_file = csv.writer(write_file) 
        
        for launch in data: # loop through each launchg 
            if (not launch['upcoming']): # only look at launches that have happened
                launch_features = []
                # the following is just extracting the features that I want to explore
                # from each launch as there is lots of information available and I dont want it all
                if (launch['rocket']['rocket_id'] == "falcon9"):
                    launch_features.append(str(launch['flight_number']))
                    launch_features.append(str(launch['launch_year']))
                    launch_features.append(str(launch['launch_date_utc']))
                    launch_features.append(str(launch['launch_success']))
                    launch_features.append(str(launch['rocket']['first_stage']['cores'][0]['flight']))
                    launch_features.append(str(launch['rocket']['second_stage']['payloads'][0]['payload_mass_kg']))
                    launch_features.append(str(launch['rocket']['second_stage']['payloads'][0]['orbit_params']['periapsis_km']))
                    launch_features.append(str(launch['rocket']['second_stage']['payloads'][0]['orbit_params']['apoapsis_km']))
                    launch_features.append(str(launch['rocket']['second_stage']['payloads'][0]['orbit_params']['inclination_deg']))
                    launch_features.append(str(launch['rocket']['second_stage']['payloads'][0]['orbit_params']['period_min']))
                    launch_features.append(str(launch['rocket']['first_stage']['cores'][0]['reused']))
                    launch_features.append(str(launch['rocket']['first_stage']['cores'][0]['landing_intent']))
                    launch_features.append(str(launch['rocket']['first_stage']['cores'][0]['land_success']))
                    launch_features.append(str(launch['launch_site']['site_name_long']))
                # all the features I wanted are extracted into an array which the csv writer uses and converts 
                # to a comma seperated line for us
                csv_file.writerow(launch_features)
        write_file.close()
```

I created the functions I will use to collect data from a get request and then parse the data and store in a file. 
basically I just get the data as json object which can be treated exactly like a json objevt can be treated in javascript. so I just iterate over every element in all the historical spacex data and get the launches that were from the falcon 9 family.

```python
#just to know that the script is running

print("Data collection started")
create_csv_file(SPACEX_DATA_FILE, headers)

print("Gathering Data...")
data = get_data_from_url(API_URL)

print("Writing Data to file...")
write_json_to_csv(SPACEX_DATA_FILE, data)

print("Data collection finished")
print("Data stored in " + SPACEX_DATA_FILE)
```

## 3. Data preparation and analysis:

## 3.A) Data preparation

```python
spacex_data = pd.read_csv(SPACEX_DATA_FILE) #read in data from csv  

pd.options.mode.chained_assignment = None  # default='warn' 
## suppress warning regarding A value is trying to be set on a copy of a slice from a DataFrame.
```

```python
#convert date from whole UTC to yyyy-mm-dd
spacex_data.launch_date = spacex_data.launch_date.map(lambda date : date[0:10])
```

Now we will remove all appearances of "None" in the data as pandas can not deal with "None"

```python
#if landing success is none assume it was a failure
spacex_data.land_success = spacex_data.land_success.map(lambda val: False if val == "None" else bool(val))

#here we can also reset values back to floats as the comparison with 'None' turns value to str
spacex_data.payload_mass = spacex_data.payload_mass.map(lambda mass: np.NaN if mass == "None" else float(mass))
spacex_data.inclination = spacex_data.inclination.map(lambda incl: np.NaN if incl == "None" else float(incl))
spacex_data.apoapsis = spacex_data.apoapsis.map(lambda app: np.NaN if app == "None" else float(app))
spacex_data.periapsis = spacex_data.periapsis.map(lambda per: np.NaN if per == "None" else float(per))
spacex_data.orbital_period = spacex_data.orbital_period.map(lambda per: np.NaN if per == "None" else float(per))
```

Now we will clean the data fully. I dont want to have any NaN values in my data so i am going to fix that.  
I will just replace all NaN values in payload_mass inclanation, apoapsis, periapsis and orbital_period with the mean values from those columns.  

```python
spacex_data = spacex_data.fillna(spacex_data.mean())
```

### Data is fully cleaned by this point

```python
#gather the number of successful and failed launches from the main data

#successful launches series
success_launch = spacex_data[spacex_data["launch_success"] == True]["launch_year"].value_counts()
#failed launches series
failed_launch = spacex_data[spacex_data["launch_success"] == False]["launch_year"].value_counts()
#augment the two series into a new dataframe
launches = pd.concat([success_launch, failed_launch], axis=1)


#gather the number of successful and failed landings from the main data
#retreive a series of successful landings
landed = spacex_data[(spacex_data.landing_intent == True) & (spacex_data.land_success == True)]["launch_year"].value_counts()
#retrieve a series of failed landings
crashed = spacex_data[(spacex_data.landing_intent == True) & (spacex_data.land_success == False)]["launch_year"].value_counts()
#put that data into a new dataframe
landings = pd.concat([landed, crashed], axis=1)

#gather all launch attempts and expendable coutn also
all_launches = spacex_data["launch_year"].value_counts()
expendable = spacex_data[spacex_data.landing_intent == False]["launch_year"].value_counts()



#new dataframe to see all launch data we have
all_launch_data = pd.concat([all_launches, expendable, landed, crashed], axis=1)


#fix headings
launches.columns = ['success', 'failure'] 
landings.columns = ['success', 'failure']
all_launch_data.columns = ['attempted', "expendable", "landed", "crashed"]

#clean any NaN values 
launches = launches.fillna(0) 
landings = landings.fillna(0)
all_launch_data = all_launch_data.fillna(0)
```

All launch data is now fully broken up into parts I want to look at. also a small bit of fixing is needed.

```python
orbital_data = spacex_data[["payload_mass","apoapsis", "periapsis", "inclination", "orbital_period"]]

small_orbits = orbital_data[(orbital_data.apoapsis < 10000) & (orbital_data.periapsis < 10000)][["apoapsis", "periapsis"]]
large_orbits = orbital_data[(orbital_data.apoapsis >= 10000) & (orbital_data.periapsis >= 10000)][["apoapsis", "periapsis"]]

mass_vs_appoaps = orbital_data[["payload_mass","apoapsis"]]
mass_vs_periaps = orbital_data[["payload_mass","periapsis"]]

mass_vs_period = orbital_data[["payload_mass","orbital_period"]]

mass_vs_inclination = orbital_data[["payload_mass","inclination"]]

orbit_vs_period = orbital_data[["apoapsis", "periapsis", "orbital_period"]]

all_orbit_features = orbital_data[["apoapsis", "periapsis", "inclination", "orbital_period"]]
```

All orbital data is now broken up into parts I believe may be interesting.

quick numerical glance of the orbital data

```python
ax4 = small_orbits.plot.scatter(x="apoapsis", y="periapsis", title="Small Orbits", figsize=(14, 8))
ax4.set_xlabel("apoapsis (km)")
ax4.set_xlabel("apoapsis (km)")

ax5 = large_orbits.plot.scatter(x="apoapsis", y="periapsis", title="Large Orbits", figsize=(14, 8))
ax5.set_xlabel("apoapsis (km)")
ax5.set_xlabel("apoapsis (km)")
print("figure 2.1 & figure 2.2")
```

SpaceX usually launch payloads int Low Earth Orbit (LEO) as we can see above. I think the most obvious reason would be due to the reusability of the First Stage Falcon 9 they can only get payloads so far up while keeping enough fuel to land.  
another interesting fact we learn from these two plots is that most orbits usually have the same apoapsis and periapsis size, meaning, that the orbits are circular and not oval shaped.

```python
ax6 = mass_vs_appoaps.plot(kind='area', stacked=False, title="Mass vs Apoapsis", color=['r', 'y'], figsize=(10, 6))
ax6.set_xlabel("launch")
ax6.set_ylabel("apoapsis (km) / mass (kg)")

ax7 = mass_vs_periaps.plot(kind='line', title="Mass vs Periapsis", color=['r', 'y'], figsize=(10, 6))
ax7.set_xlabel("launch")
ax7.set_ylabel("periapsis (km) / mass (kg)")
print("figure 2.3 & figure 2.4")
```

from figure 2.3 and figure 2.4 we can see Payload mass has very little effect on the apoapsis or periapsis. 

```python
ax8 = mass_vs_period.plot(kind='line',title="Mass vs Orbital Period", color=['m', 'c'], figsize=(10, 6))
ax8.set_xlabel("launch")
ax8.set_ylabel("period (minutes) / mass (kg)")
print("figure 2.5")
```

judging by figure 2.5 we cant conclusively say that the mass has any effect on the orbital period. there is a strange anomaly where a small payload mass has big orbital period. Howver this rule doesnt hold anywhere else. 

```python
ax8 = mass_vs_inclination.plot(subplots="true", kind='line',title="Mass vs Inclination of Orbit", color=['k', 'k'], figsize=(14, 8))
ax8[0].set_ylabel("mass (kg)")
ax8[1].set_ylabel("degrees off equator")
ax8[1].set_xlabel("launch")
print("figure 2.6")
```

when we look at the comparison of mass vs inclination of orbit we cannot see any distinct correlation so can safely say the mass of the orbit doesn't effect the inclination. 

```python
ax13 = all_orbit_features.plot(kind='area', stacked=False, title="Orbital Data", figsize=(14, 8))
ax13.set_ylabel("kilometers / degrees / minutes")
ax13.set_xlabel("launch")
ax9 = all_orbit_features.plot(subplots="true", figsize=(14,20))
ax9[0].set_title("launch apoapsis'")
ax9[1].set_title("launch perapsis'")
ax9[2].set_title("launch incalinations")
ax9[3].set_title("launch orbital periods")
ax9[0].set_ylabel("kilometers")
ax9[1].set_ylabel("kilometers'")
ax9[2].set_ylabel("degrees off equator")
ax9[3].set_ylabel("minutes")
ax9[3].set_xlabel("launch number")
print("figure 2.7 & figure 2.8")
```

Looking at all the orbital data in one chart (figure 2.7) is a bit confusing because the different ranges. so we break all the data up into separate diagrams so we can inspect it side by side (figure 2.8). 

Again we can see that apoapsis and periapsis are usually always very close in size meaning most orbits are circular but there was one orbit which has a much bigger apoapsis meaning that it has a very oblong shape.
  
On a final look at all orbital data together we see clearly and can conclude that the size of the orbit plays a huge part in the the length of the orbital period. Also we can see that the inclination of the orbit is not tied to any other variable which makes sense.
